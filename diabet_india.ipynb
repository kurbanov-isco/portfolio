```python
from catboost import CatBoostClassifier, Pool, cv
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss, confusion_matrix, accuracy_score
```

```python
train, test = train_test_split(df, train_size=0.6, random_state=42, stratify=df['Outcome'], shuffle=True)
val, test = train_test_split(test, train_size=0.5, random_state=42,stratify=test['Outcome'], shuffle=True)
```

```python
X = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']
y = ['Outcome']
```

```python
from catboost import Pool

train_data = Pool(data=train[X],
                  label=train[y]
                                 )

valid_data = Pool(data=val[X],
                  label=val[y]
                                   )

test_data = Pool(data=test[X],
                  label=test[y]
                                )
```

```python
params = {'verbose':100,
          'eval_metric':'AUC',
          'loss_function': 'Logloss',
          'random_seed':42,
          'learning_rate':0.05}
```

```python
model = CatBoostClassifier(**params)
```

```python
model.fit(train_data, eval_set= valid_data)
```

```python
test['predict'] = model.predict_proba(test[X])[:,1]
```

```python
test.sort_values(by='predict',ascending=False).head(round(len(test)*0.2))['Outcome'].sum() / round(test['Outcome'].sum())
```

```python
result = []
for thr in tqdm(thrs):
  test['thrs_pred'] = (test['predict'] > thr) *1
  result.append((thr,
                 f1_score(test['Outcome'],test['thrs_pred'])))
```

```python
def metricsclass(y_true, y_pred):
  print(recall_score(y_true, y_pred))
  print(precision_score(y_true, y_pred))
  print(f1_score(y_true, y_pred))
  print(roc_auc_score(y_true, y_pred))
```
